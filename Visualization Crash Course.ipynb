### Style calls

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

- Limit of axis

```python
plt.plot(x,y)

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

- Change color

```python
plt.plot(x,y,color='red')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='#bc67c9')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

- Add marker

```python
plt.plot(x,y,color='#bc67c9', marker='*', markersize=15, linestyle='--')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

### Distribution plots (Histograms)

```python
sns.distplot(df['age'])
```

```python
sns.distplot(df['age'], kde=False)
```

```python
sns.distplot(df['age'], kde=False, bins=2)
```

```python
sns.distplot(df['age'], kde=False, bins=50)
```

### Change the size of the plot

```python
plt.figure(figsize=(8,4))
sns.distplot(df['age'], kde=False, bins=50)
```

### Change the color of the plot

```python
plt.figure(figsize=(8,4))
sns.distplot(df['age'], kde=False, bins=50, color='red')
```

### Box plots
They show you the distribution and specifically the distribution across different __categorical features__. 
![Box plots](img/boxplot.png)


```python
#x shoud be a categorical feature and y should be a continous one.
sns.boxplot(x='sex', y='age', data=df)
```

```python
sns.boxplot(x='target', y='thalach', data=df)
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```
